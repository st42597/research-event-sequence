```python
import nltk
import pandas as pd
import json
import re
import os
import copy
nltk.download('book')
```

```python
wine130_df = pd.read_csv('wine130.csv')
wine150_df = pd.read_csv('wine150.csv')
```

```python
if os.path.isfile('./wine_df.json'):
    with open('wine_df.json', 'r') as json_file:
        wine_df = json.load(json_file)
else:
    wine_df = pd.concat([wine130_df['description'], wine150_df['description']]).reset_index(drop=True)
    wine_df = list(map(lambda x: x.replace('\n', ' '), wine_df))
    wine_df = list(map(lambda x: re.sub(r'[^ A-Za-z]', '', x), wine_df))
    wine_df = list(map(lambda x: nltk.tokenize.word_tokenize(x), wine_df))
    wine_df = list(map(lambda x: list(map(lambda y: y.lower(), x)), wine_df))
    wine_df = list(map(lambda x: nltk.tag.pos_tag(x), wine_df))
    
    json_value = json.dumps(wine_df)
    json_value = json.loads(json_value)
    with open('wine_df.json', 'w') as json_file:
        json.dump(json_value, json_file, indent=2)

wine_df
```

```python
word2event = {
    'fruit': 'F',
    'fruity': 'F',
    'aromas': 'A',
    'apple': 'AP',
    'berry': 'B',
    'blackberry': 'B',
    'raspberry': 'B',
    'cherry': 'C',
    'tannins': 'T',
    'acidity': 'AC',
    'oak': 'O',
    'cabernet': 'CA',
    'spice': 'S',
    'plum': 'P',
    'ripe': 'RR',
    'blend': 'BL',
    'note': 'N',
    'notes': 'N',
    'dry': 'D',
    'sweet': 'SW',
    'fresh': 'F',
    'soft': 'SO',
    'vanilla': 'V',
    'chocolate': 'CH',
}
'''
'pepper': 'PE',
'peach': "PEA"
'''
```

```python
filtered = list(map(lambda x: list(filter(lambda y: y[0] in word2event, x)), wine_df))
filtered = list(filter(lambda x: len(x) >= 8 , filtered))
filtered = list(map(lambda x: x[:8], filtered))
filtered = list(map(lambda x: list(map(lambda y: '#' + word2event[y[0]], x)), filtered))
filtered = list(map(lambda x: {'es': x, 'order': -1}, filtered))
for i in range(len(filtered)):
    filtered[i]['id'] = i
print(len(filtered))
```

```python
AoI_MAX_LENGTH = 2
cnt = {}
for j in range(len(filtered[0]['es']) - 1):
  for i in range(len(filtered)):
    for k in range(2, AoI_MAX_LENGTH + 1):
      if j + k - 1 >= len(filtered[0]['es']): break
      AoI = filtered[i]['es'][j] + filtered[i]['es'][j + k - 1] + str(j) + '-' + str(j + k - 1)
      if AoI in cnt:
        cnt[AoI] = [cnt[AoI][0] + 1, cnt[AoI][1], cnt[AoI][2] | {i}]
      else:
        cnt[AoI] = [1, j, set([i])]
AoI_count = list(cnt.items())
AoI_count = sorted(AoI_count, key=lambda x: -x[1][0])

print(AoI_count)

sorted_filtered = list([*filtered])

# print(sorted_filtered)
```

```python
# threshold  node 
tmp = []
for i in range(len(AoI_count)):
    if AoI_count[i][1][0] >= 10:
        tmp.append(AoI_count[i])
AoI_count = tmp

len(AoI_count)
```

```python
# node    node  distance 
def jaccard_similarity(x, y):
    intersection_cardinality = len(set.intersection(*[x, y]))
    union_cardinality = len(set.union(*[x, y]))
    return intersection_cardinality / float(union_cardinality)

num_nodes = len(AoI_count)
dist = [[0 for i in range(num_nodes)] for j in range(num_nodes)]
for i in range(num_nodes):
    for j in range(num_nodes):
        dist[i][j] += jaccard_similarity(AoI_count[i][1][2], AoI_count[j][1][2])
```

```python
# distance  node 

NODE = 0

dist[NODE][NODE] = -1
for i in range(num_nodes):
    AoI_count[i][1].append(dist[NODE][i])
sorted_AoI_count = sorted(AoI_count, key=lambda x: x[1][3])

for i in range(len(sorted_filtered)):
    sorted_filtered[i]['order'] = 0
for i in range(len(sorted_AoI_count)):
  for j in range(len(sorted_filtered)):
    if sorted_filtered[j]['id'] in sorted_AoI_count[i][1][2]:
      sorted_filtered[j]['order'] += 2**(len(sorted_AoI_count) - i)

sorted_filtered = sorted(sorted_filtered, key=lambda x: -x['order'])
```

```python
tmp = []
for i in range(20):
    for j in AoI_count[i][1][2]:
        tmp.append(copy.deepcopy(sorted_filtered[j]))
sorted_filtered = tmp
```

```python
color = [
    "#003f5c", "#2f4b7c", "#665191", "#a05195", "#d45087",
    "#f95d6a", "#ff7c43", "#ffa600", "#488f31", "#28b4c8",
    "#a2c8ec", "#FFD700", "#87CEEB", "#FFB6C1", "#20B2AA",
    "#778899", "#32CD32", "#6A5ACD", "#ff6347", "#90ee90"
]
event = set(['#' + word2event[key] for key in word2event])
cnt = 0
event2color = {}
for key in event:
    event2color[key] = color[cnt]
    cnt += 1

result = []
for i in range(len(sorted_filtered)):
    result.append(sorted_filtered[i]['es'])
for i in range(len(result)):
    for j in range(len(result[i])):
        result[i][j] = event2color[result[i][j]] 
print(result)
```

```python
dx = [1, 0, 0, -1]
dy = [0, 1, -1, 0]

cnt = 0
total = 0
for i in range(len(result)):
    for j in range(len(result[i])):
        for k in range(4):
            nx = i + dx[k]
            ny = j + dy[k]
            if nx < 0 or nx >= len(result) or ny < 0 or ny >= len(result[i]): continue
            total += 1
            if result[i][j] == result[nx][ny]: cnt += 1

print(cnt / total)
```

```python
label = []
for key in word2event:
    label.append([key, event2color['#' + word2event[key]]])
lst = [label, result]
json_value = json.dumps(lst)
json_value = json.loads(json_value)
with open('wine.json', 'w') as json_file:
    json.dump(json_value, json_file, indent=2)
```
