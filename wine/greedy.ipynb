```python
import nltk
import pandas as pd
import json
import re
nltk.download('book')
```

```python
wine130_df = pd.read_csv('wine130.csv')
wine150_df = pd.read_csv('wine150.csv')
```

```python
wine_df = pd.concat([wine130_df['description'], wine150_df['description']]).reset_index(drop=True)
wine_df = list(map(lambda x: x.replace('\n', ' '), wine_df))
wine_df = list(map(lambda x: re.sub(r'[^ A-Za-z]', '', x), wine_df))
wine_df = list(map(lambda x: nltk.tokenize.word_tokenize(x), wine_df))
wine_df = list(map(lambda x: nltk.tag.pos_tag(x), wine_df))
wine_df
```

```python
cnt = {}
for i in range(len(wine_df)):
  for j in range(len(wine_df[i])):
    if wine_df[i][j][0] in cnt:
      cnt[wine_df[i][j][0]] += 1
    else:
      cnt[wine_df[i][j][0]] = 1
words_count = list(cnt.items())
words_count = sorted(words_count, key=lambda x: -x[1])
print(words_count)
```

```python
word2event = {
    'fruit': 'F',
    'fruity': 'F',
    'Aromas': 'A',
    'aromas': 'A',
    'apple': 'AP',
    'berry': 'B',
    'blackberry': 'B',
    'raspberry': 'B',
    'cherry': 'C',
    'tannins': 'T',
    'acidity': 'AC',
    'oak': 'O',
    'Cabernet': 'CA',
    'spice': 'S',
    'plum': 'P',
    'ripe': 'RR',
    'blend': 'BL',
    'note': 'N',
    'notes': 'N',
    'dry': 'D',
    'sweet': 'SW',
    'fresh': 'F',
    'soft': 'SO',
    'vanilla': 'V',
    'chocolate': 'CH',
}
'''
'pepper': 'PE',
'peach': "PEA"
'''
```

```python
filtered = list(map(lambda x: list(filter(lambda y: y[0] in word2event, x)), wine_df))
filtered = list(filter(lambda x: len(x) >= 8 , filtered))
filtered = list(map(lambda x: x[:8], filtered))
filtered = list(map(lambda x: list(map(lambda y: '#' + word2event[y[0]], x)), filtered))
filtered = list(map(lambda x: {'es': x, 'order': -1}, filtered))
for i in range(len(filtered)):
    filtered[i]['id'] = i
print(len(filtered))
```

```python
AoI_MAX_LENGTH = 4
cnt = {}
for j in range(len(filtered[0]['es']) - 1):
  for i in range(len(filtered)):
    for k in range(2, AoI_MAX_LENGTH + 1):
      if j + k - 1 >= len(filtered[0]['es']): break
      AoI = filtered[i]['es'][j] + filtered[i]['es'][j + k - 1] + str(j) + '-' + str(j + k - 1)
      if AoI in cnt:
        cnt[AoI] = [cnt[AoI][0] + 1, cnt[AoI][1], cnt[AoI][2] | {i}]
      else:
        cnt[AoI] = [1, j, set([i])]
AoI_count = list(cnt.items())
AoI_count = sorted(AoI_count, key=lambda x: -x[1][0])

print(AoI_count)

sorted_filtered = list([*filtered])

num = 1
for i in range(len(AoI_count) - 1, -1, -1):
  for j in range(len(sorted_filtered)):
    if sorted_filtered[j]['id'] in AoI_count[i][1][2]:
      sorted_filtered[j]['order'] += num
  num = num * 2

sorted_filtered = sorted(sorted_filtered, key=lambda x: -x['order'])

# print(sorted_filtered)
```

```python
color = [
    "#1f77b4",
    "#ff7f0e",
    "#2ca02c",
    "#d62728",
    "#9467bd",
    "#8c564b",
    "#e377c2",
    "#7f7f7f",
    "#bcbd22",
    "#17becf",
    "#aec7e8",
    "#ffbb78",
    "#98df8a",
    "#ff9896",
    "#c5b0d5",
    "#c49c94",
    "#f7b6d2",
    "#c7c7c7",
    "#dbdb8d",
    "#9edae5",
]

cnt = 0
event2color = {}
result = []
for i in range(len(sorted_filtered)):
    result.append(sorted_filtered[i]['es'])
for i in range(len(result)):
    for j in range(len(result[i])):
        if result[i][j] not in event2color:
            event2color[result[i][j]] = color[cnt]
            # print(result[i][j])
            cnt += 1
        result[i][j] = event2color[result[i][j]] 
print(result)
```

```python

dx = [1, 0, 0, -1]
dy = [0, 1, -1, 0]

cnt = 0
total = 0
for i in range(len(result)):
    for j in range(len(result[i])):
        for k in range(4):
            nx = i + dx[k]
            ny = j + dy[k]
            if nx < 0 or nx >= len(result) or ny < 0 or ny >= len(result[i]): continue
            total += 1
            if result[i][j] == result[nx][ny]: cnt += 1

print(cnt / total)
```

```python
label = []
for key in word2event:
    label.append([key, event2color['#' + word2event[key]]])
lst = [label, result]
json_value = json.dumps(lst)
json_value = json.loads(json_value)
with open('wine.json', 'w') as json_file:
    json.dump(json_value, json_file, indent=2)
```
